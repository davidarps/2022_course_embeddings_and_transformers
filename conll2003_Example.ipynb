```python
# write the dataset tokens to file
for ds in ['train', 'validation']:
    with open('data/' + ds+'.tok', 'w') as f:
        for sent in datasets[ds]['tokens'][:cutoff]:
            f.write(' '.join(sent))
            f.write('\n')
```

```python
# decode the pos tags from the dataset and write to file
for ds in ['train', 'validation']:
    with open('data/'+ds+'.pos', 'w') as f:
        for sent in datasets[ds]['pos_tags'][:cutoff]:
            f.write(' '.join([datasets[ds].features['pos_tags'].feature.names[t] for t in sent]))
            f.write('\n')
```

```python
import neurox.interpretation.utils as utils
X, y, mapping = utils.create_tensors(tokens, activations, 'NN', x_dtype='float16')
label2idx, idx2label, src2idx, idx2src = mapping
```

# Train Probing Classifier

```python
import neurox.interpretation.linear_probe as linear_probe
probe = linear_probe.train_logistic_regression_probe(X, y, lambda_l1=0.001, lambda_l2=0.001)
```

# Evaluate Model

# Task 5

```python
linear_probe.evaluate_probe(probe, X, y, idx_to_class=idx2label)
```
